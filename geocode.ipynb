```python
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
```

```python
user_agent = "GigsVis/1.0 (@hollowscene)"
geolocator = Nominatim(user_agent=user_agent)
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1, max_retries=3)
```

```python
csv_path = "raw_data/gig_log-20260611.csv"
df = pd.read_csv(csv_path)
```

```python
df["coordinates"] = df["Street Address"].apply(geocode)
df["latitude"] = df["coordinates"].apply(lambda loc: loc.latitude if loc else None)
df["longitude"] = df["coordinates"].apply(lambda loc: loc.longitude if loc else None)
```

```python
columns = [
    "Headliner",
    "Opener 1",
    "Opener 2",
    "Opener 3",
    "Date",
    "Day of week",
    "Start time (doors open)",
    "Gig Location",
    "Street Address",
    "Headliner from",
    "Genre",
    "latitude",
    "longitude",
]

column_renames = {
    "Headliner": "headliner",
    "Opener 1": "opener1",
    "Opener 2": "opener2",
    "Opener 3": "opener3",
    "Date": "date",
    "Day of week": "day_of_week",
    "Start time (doors open)": "start_time",
    "Gig Location": "location_name",
    "Street Address": "location_address",
    "Headliner from": "headliner_from",
    "Genre": "genre",
    "latitude": "location_lat",
    "longitude": "location_long",
}

df = df[columns].rename(columns=column_renames)
df
```

```python
df.to_csv("processed_data/gig_log.csv", index=False)
```
