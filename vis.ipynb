```python
import branca
import folium
import pandas as pd
```

```python
csv_path = "processed_data/gig_log.csv"
df = pd.read_csv(csv_path)
df
```

```python
m = folium.Map(
    location=(-37.814012219263226, 144.9601751713214),
    control_scale=True,
    zoom_start=13,
)

for index, row in df.iterrows():
    headliner = row["headliner"]
    date = row["date"]
    address = row["location_address"]
    location_name = row["location_name"]
    location_lat = row["location_lat"]
    location_long = row["location_long"]

    html = f"""
    <body style="font-family: sans-serif">
    <h1>{headliner}</h1>
    <h2>{location_name}</h2>
    """

    iframe = branca.element.IFrame(html=html, width=500, height=300)
    popup = folium.Popup(iframe, max_width=500)

    if pd.isna(location_lat) or pd.isna(location_long):
        print(f"Skipped marker for address `{address}`.")
        continue

    folium.Marker(
        location=[location_lat, location_long],
        tooltip=date,
        popup=popup,
        lazy=True,
    ).add_to(m)
    print(f"Added marker for address `{address}`.")
```
