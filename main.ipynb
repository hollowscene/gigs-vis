```python
import branca
import folium
import pandas as pd
from geopy.geocoders import Nominatim
```

```python
user_agent = "GigsVis/1.0 (https://github.com/hollowscene)"
geolocator = Nominatim(user_agent=user_agent)
```

```python
df = pd.read_csv("gig_log-20241221.csv")
df
```

```python
m = folium.Map(
    location=(-37.814012219263226, 144.9601751713214),
    control_scale=True,
    zoom_start=13,
)
```

```python
for index, row in df.iterrows():
    headliner = row["Headliner"]
    date = row["Date"]
    address = row["Street Address"]
    location_name = row["Gig Location"]
    coordinates = geolocator.geocode(address)

    if coordinates is None:
        print(f"Address `{address}` could not be geolocated")
        continue

    html = f"""
    <body style="font-family: sans-serif">
    <h1>{headliner}</h1>
    <h2>{location_name}</h2>
    """

    iframe = branca.element.IFrame(html=html, width=500, height=300)
    popup = folium.Popup(iframe, max_width=500)

    folium.Marker(
        location=[coordinates.latitude, coordinates.longitude],
        tooltip=date,
        popup=popup,
        lazy=True,
    ).add_to(m)
    print(f"Added marker for address `{address}`")
```
