```python
import folium
import pandas as pd
from geopy.geocoders import Nominatim
```

```python
user_agent = "GigsVis/1.0 (https://github.com/hollowscene)"
geolocator = Nominatim(user_agent=user_agent)
```

```python
df = pd.read_csv("gig_log-20241221.csv")
df
```

```python
m = folium.Map(location=(-37.814012219263226, 144.9601751713214), zoom_start=13)
```

```python
for index, row in df.iterrows():
    headliner = row["Headliner"]
    date = row["Date"]
    address = row["Street Address"]
    location = geolocator.geocode(address)

    if location is None:
        print(f"Address `{address}` could not be geolocated")

    else:
        folium.Marker(
            location=[location.latitude, location.longitude],
            tooltip=headliner,
            popup=date,
        ).add_to(m)
        print(f"Added marker for address `{address}`")
```
